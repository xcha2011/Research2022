```python
import pandas as pd
import re
```

```python
# data from the umd website
cr = pd.read_csv("MSR Postings.csv")

cr.head()
```

### Order by the listing number

```python
ordered_cr = cr.sort_values(by='Listing Number')

ordered_cr.head()
```

```python
ordered_cr = ordered_cr.set_index('Listing Number')

ordered_cr.head()
```

### Find all rows with python in 'Keywords' column or 'Required Skills' column

```python
# find the indexes that I am looking for
filtered_rows = []

for index, row in ordered_cr.iterrows():
    if re.search('(P|p)ython', str(row['Keywords'])) != None or re.search('(P|p)ython', str(row['Required Skills'])) != None:
        filtered_rows.append(index)
        
filtered_rows
```

```python
filtered_cr = ordered_cr.filter(items = filtered_rows, axis=0)

filtered_cr
```

### Creating a Match Score

```python
# Things that I am looking for in research (other than python)
interests = [
    '(P|p)rogram([a-r]|[t-z])+', # Covers programming and programmer, avoids program(s) (these mean something different) 
    '(M|m)achine (L|l)earning', 
    '(Q|q)uantum',
    '(AI|A I|(A|a)rtificial (I|i)ntelligence)',
]
```

```python
scores = []
for index, row in filtered_cr.iterrows():
    count = 0
    for interest in interests:
        a = False
        for string in row:
            string = str(string)
            if re.search(interest, string) != None:
                a = True
            if a:
                break
        if a:
            count += 1
    scores.append(count/len(interests))

filtered_cr['Score'] = scores 
```

```python
scored_cr = filtered_cr.sort_values(by='Score', ascending=False)
scored_cr = scored_cr[(scored_cr['Score'] >= .5)]  
```
